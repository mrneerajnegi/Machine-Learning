```python
import pandas as pd
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import SCORERS
```

```python
df=pd.read_csv("./Advertising.csv")
df.head()
X=df.drop("sales",axis=1)
y=df["sales"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
scaler_model=StandardScaler()
scaler_model.fit(X_train)
X_train=scaler_model.transform(X_train)
X_test=scaler_model.transform(X_test)
```

```python
elastic_model=ElasticNet()
params={"alpha":[0.1,0.2,0.5,1],"l1_ratio":[0.1,0.2,0.5,1]}
grid_model=GridSearchCV(estimator=elastic_model,scoring="neg_mean_absolute_error",
                        cv=5,param_grid=params)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
grid_model.best_estimator_
```

```python
results=grid_model.cv_results_
df=pd.DataFrame(results)
df
```

```python
prediction=grid_model.predict(X_test)
prediction
```
