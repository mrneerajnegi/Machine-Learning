```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("./hearing_test.csv")
X=df.drop("test_result",axis=1)
Y=df["test_result"]
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.10, random_state=42)
standard_model=StandardScaler()
X_train=standard_model.fit_transform(X_train)
X_test=standard_model.fit_transform(X_test)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
prediction=model.predict(X_test)
```

```python
prediction
```

```python
model.predict_proba(X_test)
```

# confusion_matrix  ,  plot_confusion_matrix

```python
from sklearn.metrics import accuracy_score,confusion_matrix,plot_confusion_matrix
```

```python
accuracy_score(y_test,prediction)
```

```python
confusion_matrix(y_test,prediction)
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
fig,ax=plt.subplots(figsize=(5,5),dpi=200)
plot_confusion_matrix(model,X_test,y_test,normalize='true',ax=ax)
```

# classification_report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction,))
```

# precision_score  ,  recall_score

```python
from sklearn.metrics import precision_score,recall_score
```

```python
precision_score(y_test,prediction,)
```

```python
recall_score(y_test,prediction)
```

# plot_roc_curve , plot_precision_recall_curve

```python
from sklearn.metrics import plot_roc_curve,plot_precision_recall_curve
```

```python
fig,ax=plt.subplots(figsize=(6,4),dpi=200)

plot_roc_curve(model,X_test,y_test,ax=ax)
```

```python
fig,ax=plt.subplots(figsize=(6,4),dpi=200)

plot_precision_recall_curve(model,X_test,y_test,ax=ax)
```
