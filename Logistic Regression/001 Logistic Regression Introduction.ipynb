```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df=pd.read_csv("./hearing_test.csv")
df.head()
```

```python
df["test_result"].value_counts()
```

```python
sbn.countplot(data=df,x="test_result")
```

```python
plt.figure(figsize=(8,4),dpi=200)
sbn.boxplot(data=df,x="test_result", y="physical_score")
```

```python
plt.figure(figsize=(8,4),dpi=200)
sbn.boxplot(data=df,x="test_result", y="age")
```

```python
plt.figure(figsize=(8,4),dpi=200)
sbn.scatterplot(x="age",y="physical_score",data=df,hue="test_result",alpha=0.5)
```

```python
plt.figure(figsize=(8,4),dpi=200)
sbn.pairplot(data=df,hue="test_result")
```
