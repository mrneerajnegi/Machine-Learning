```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.linear_model import LogisticRegression,LogisticRegressionCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,plot_confusion_matrix,classification_report
```

```python
df =pd.read_csv("./iris.csv")
df.head()
```

```python
df["species"].value_counts()
```

```python
sbn.scatterplot(x="sepal_length",y="petal_length",data=df,hue="species")
```

```python
sbn.countplot(x="species",data=df)
```

```python
X=df.drop("species",axis=1)
Y=df["species"]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
standard=StandardScaler()
X_train=standard.fit_transform(X_train)
X_test=standard.transform(X_test)
```

```python
model=LogisticRegression(multi_class="ovr",solver="saga",max_iter=5000)
params={"penalty":["l1","l2","elasticnet"],"l1_ratio":np.linspace(0,1)}
gridcv=GridSearchCV(model,param_grid=params,)
gridcv.fit(X_train,y_train)
```

```python
gridcv.best_params_
```

```python
prediction=gridcv.predict(X_test)
prediction
```

```python
accuracy_score(y_test,prediction)
```

```python
plot_confusion_matrix(gridcv,X_test,y_test,normalize="true")
```

```python
print(classification_report(prediction,y_test))
```

```python
coefs=np.sort(gridcv.best_estimator_.coef_[0])
coefs
```

```python
sbn.barplot(y=coefs,x=X.columns)
```
