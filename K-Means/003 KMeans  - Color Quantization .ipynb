```python
- The image is a 3 D array (H , W, C)
- Each Pixxel has an RGB value to create a color
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.image as plt_image
import numpy as np
```

```python
image_array_3d=plt_image.imread("./palm-trees.jpg")
image_array_3d.shape   #(H,W,C)
```

```python
plt.figure(dpi=200)
plt.imshow(image_array_3d);
```

```python
#(H,W,C) ------> 2D  (H*W , C) 
#because kMeans only deals with 2D
```

```python
# name=np.array([[[1,2,3,4],[1,2,3,4]]])
# print(name.shape)
# name.reshape(2,4)
```

```python
(h,w,c)=image_array_3d.shape
```

```python
image_array_2d=image_array_3d.reshape(h*w,c)
```

```python
image_array_2d.shape
```

```python
kmodel=KMeans(n_clusters=6)
cluster_points=kmodel.fit_predict(image_array_2d)
```

```python
#Number of colors
print(cluster_points)
```

```python
#these are the 6 colors used, these are basically center of each clusters i.e RGB
rgb_points= kmodel.cluster_centers_
rgb_points
```

```python
image_3d_=rgb_points[cluster_points].round(0).astype(int)
image_3d_
```

```python
image_3d_=image_3d_.reshape(h,w,c)
```

```python
plt.figure(dpi=200)
plt.imshow(image_3d_);
```
