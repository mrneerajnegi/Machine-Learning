```python
import numpy as np
import matplotlib.pyplot as plt

#for old versions where images doesnt show
%matplotlib inline
```

```python
B=np.linspace(0,10,11)
X=np.arange(0,11)
Y=X**4
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
```

```python
#Move
fig=plt.figure()
axes=fig.add_axes([0,0,0.5,0.5])

```

```python
#Plot
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,Y)
```

```python
#Plot
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,B)

```

```python
#add other figure as a zoom 
fig1=plt.figure()
axes=fig1.add_axes([0,0,1,1])
axes.plot(X,Y)
axes.set_xlabel("X axis")
axes.set_ylabel("Y Axis")
axes.set_xlim(0,9)
axes.set_ylim(0,7000)

#Small Axis
fig2=plt.figure()
axes=fig1.add_axes([0.2,0.2,0.3,0.3])
axes.plot(X,Y)
axes.set_xlabel("X axis")
axes.set_ylabel("X axis")
axes.set_title("Zoom")
axes.set_xlim(3,4)
axes.set_ylim(0,500)
```
