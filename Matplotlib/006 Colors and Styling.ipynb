```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(1,10)
X
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3,ls="--") #or linestyle
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3,ls="--",marker="o") 
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3,ls="--",marker="o", ms="10") #or markerstyle
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3,label="X vs X**3",ls="--",marker="o", ms="10") 
axes.plot(X,X+2,label="X vs X+2",ls="--",marker="o", ms="10")
axes.legend()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3,label="X vs X**3",ls="--",marker="o", ms="10",color="red")  # we can mention direct color code also
axes.plot(X,X+2,label="X vs X+2",ls="--",marker="o", ms="10",color="black")
axes.legend()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3,label="X vs X**3",ls="--",marker="o", ms="10",color="red",lw=5)  #we can use linewidth also
axes.plot(X,X+2,label="X vs X+2",ls="--",marker="o", ms="10",color="black")
axes.legend()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X**3,label="X vs X**3",ls="--",marker="o", ms="10",color="red",lw=5)  #we can use linewidth also
axes.plot(X,X+2,label="X vs X+2",ls="--",marker="o", ms="10",color="black")
lines=axes.plot(X,X**4,label="X vs X**5",color="blue",lw=10)
lines[0].set_dashes([1,2,5,1]) # 1-dash , 2-black 5-dash, 1-blank
axes.legend()
```
