```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(0,10)
Y=X*2
X
```

```python
Y
```

```python
figure=plt.figure()
axes=figure.add_axes([0,0,1,1])
axes.plot(X,Y)
```

```python
# dpi for better clearity
figure=plt.figure(dpi=80)
axes=figure.add_axes([0,0,1,1])
axes.plot(X,Y)

```

```python
# dpi for better clearity
figure=plt.figure(dpi=150,figsize=(5,5))
axes=figure.add_axes([0,0,1,1])
axes.plot(X,Y)
```

```python
# dpi for better clearity
figure=plt.figure(dpi=150,figsize=(5,5))
axes=figure.add_axes([0,0,1,1])
axes.plot(X,Y)
figure.savefig("a.png")
#it will not shown x axis and y axis in saved image so we use bbox_inches=tight
```

```python
figure=plt.figure(dpi=150,figsize=(5,5))
axes=figure.add_axes([0,0,1,1])
axes.plot(X,Y)
figure.savefig("a.png",bbox_inches="tight")
```
