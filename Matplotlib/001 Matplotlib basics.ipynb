```python
import numpy as np
import matplotlib.pyplot as plt

#for old versions where images doesnt show
%matplotlib inline
```

```python
X=np.arange(0,15)
```

```python
Y=1*X
```

```python
plt.plot(X,Y)
plt.title("Graph")
plt.xlabel("X axis")
plt.ylabel("X axis")
plt.show()
```

```python
#limit x
plt.plot(X,Y)
plt.title("Graph")
plt.xlabel("X axis")
plt.ylabel("X axis")
plt.xlim(0,10)
plt.show()
```

```python
#save figure
plt.plot(X,Y)
plt.title("Graph")
plt.xlabel("X axis")
plt.ylabel("X axis")
plt.xlim(0,10)
plt.ylim(0,14)
plt.savefig("abc.png")
plt.show()
```
