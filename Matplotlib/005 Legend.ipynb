```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(1,10)
X
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X+2)
axes.plot(X,X**2)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X+2,label="X vs X+2")
axes.plot(X,X**2,label="X vs X**2")
axes.legend()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X+2,label="X vs X+2")
axes.plot(X,X**2,label="X vs X**2")
axes.legend(loc="lower left")
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X+2,label="X vs X+2")
axes.plot(X,X**2,label="X vs X**2")
axes.legend(loc=(1,1))
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(X,X+2,label="X vs X+2")
axes.plot(X,X**2,label="X vs X**2")
axes.legend(loc=(1.1,0.5))
```
