```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X= np.arange(0,11)
Y=X**4
```

```python
(fig,axes)=plt.subplots(nrows=1,ncols=2)
```

```python
(fig,axes)=plt.subplots(nrows=2,ncols=2)
```

```python
(fig,axes)=plt.subplots(nrows=3,ncols=1)
```

```python
(fig,axes)=plt.subplots(nrows=3,ncols=1)
axes.shape
```

```python
(fig,axes)=plt.subplots(nrows=1,ncols=2)
axes[0].plot(X,Y)
axes[1].plot(X,Y)
axes.shape
```

```python
(fig,axes)=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(X,Y)
axes[0][1].plot(X,X)
axes[1][0].plot(X,Y)
axes[1][1].plot(X,Y)
```

```python
(fig,axes)=plt.subplots(nrows=3,ncols=1)
axes[0].plot(X,Y)
axes[1].plot(X,X)
axes[2].plot(X,X)
```

```python
(fig,axes)=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(X,Y)
axes[0][1].plot(X,X)
axes[1][0].plot(X,Y)
axes[1][1].plot(X,Y)
plt.tight_layout() # for giving some padding in between the axes
```

```python
(fig,axes)=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(X,Y)
axes[0][0].set_xlabel("X axis")
axes[0][0].set_ylabel("Y axis")



axes[0][1].plot(X,X)
axes[1][0].plot(X,Y)
axes[1][0].set_xlabel("X axis")
axes[1][0].set_ylabel("Y axis")
axes[1][0].set_title("1st axes")


axes[1][1].plot(X,Y)
plt.tight_layout() 
```

```python
(fig,axes)=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
axes[0][0].plot(X,Y)
axes[0][0].set_xlabel("X axis")
axes[0][0].set_ylabel("Y axis")



axes[0][1].plot(X,X)
axes[1][0].plot(X,Y)
axes[1][0].set_xlabel("X axis")
axes[1][0].set_ylabel("Y axis")
axes[1][0].set_title("1st axes")
axes[1][1].plot(X,Y)

fig.suptitle("Yo Yo",fontsize=20)
fig.subplots_adjust(wspace=1,hspace=0.5) # in %
fig.savefig("avs.png")
```
