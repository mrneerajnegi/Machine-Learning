```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("../data/05-Seaborn/dm_office_sales.csv")
df
```

```python
plt.figure(figsize=(10,4),dpi=200)
sbn.countplot(x="level of education",data=df)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sbn.countplot(x="level of education",data=df)
plt.ylim(0,1000)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sbn.countplot(x="level of education",data=df,hue="division")
# plt.ylim(0,1000)
plt.legend(loc=(1.1,0.5))
```

```python
plt.figure(figsize=(10,4),dpi=200)
sbn.barplot(data=df,x="level of education",y="salary")
```

```python
plt.figure(figsize=(10,4),dpi=200)
sbn.barplot(data=df,x="level of education",y="salary",hue="training level")
plt.legend(loc=(1.1,0.5))
```

```python
plt.figure(figsize=(10,4),dpi=200)
sbn.barplot(data=df,x="level of education",y="salary",hue="training level")
plt.legend(loc=(1.1,0.5))

```

```python
# we can also aggregate method in barplot, bydefault it is mean
plt.figure(figsize=(10,4),dpi=200)
sbn.barplot(data=df,x="level of education",y="salary",hue="training level",estimator=np.mean)
plt.legend(loc=(1.1,0.5))

```
