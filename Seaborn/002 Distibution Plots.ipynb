```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np
```

```python
df=pd.read_csv("../data/05-Seaborn/dm_office_sales.csv")
df
```

```python
sbn.rugplot(x="salary",data=df,height=0.5)
```

```python
#it is more genral way to try to use this if we compare with histplot
sbn.displot(x="salary",data=df,color='orange',bins=10,edgecolor='red')
```

```python
sbn.histplot(x="salary",data=df,color='orange',bins=10,edgecolor='red',lw=3)
```

```python
#we can use kde and rug plot with displot that is not possible with histplot
sbn.displot(x="salary",data=df,color='orange',bins=10,edgecolor='red',kde=True,rug=True)
```

```python
sbn.histplot(x="salary",data=df,color='orange',bins=10,edgecolor='red',kde=true)
```

```python
sbn.kdeplot(x="salary",data=df)
```
