```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df=pd.read_csv("../data/05-Seaborn/dm_office_sales.csv")
df
```

```python
sbn.scatterplot(x="salary",y="sales",data=df)
```

```python
plt.figure(figsize=(10,6),dpi=200)
sbn.scatterplot(x="salary",y="sales",data=df)
```

```python
plt.figure(figsize=(10,6),dpi=200)
sbn.scatterplot(x="salary",y="sales",data=df,alpha=0.5)
```

```python
#hue => show based in each category, in categorical(using symbols )and in continous using different color opacity
plt.figure(figsize=(10,6),dpi=200)
sbn.scatterplot(x="salary",y="sales",data=df,alpha=0.5,hue="work experience")
```

```python
plt.figure(figsize=(10,6),dpi=200)
sbn.scatterplot(x="salary",y="sales",data=df,alpha=0.5,hue="level of education",)
```

```python
#style => give different markers
plt.figure(figsize=(10,6),dpi=200)
sbn.scatterplot(x="salary",y="sales",data=df,style="level of education",)
```

```python
#s => size
plt.figure(figsize=(10,6),dpi=200)
sbn.scatterplot(x="salary",y="sales",data=df,style="level of education",hue="division")
```

```python
plt.figure(figsize=(10,6),dpi=200)
sbn.scatterplot(x="salary",y="sales",data=df,style="level of education",hue="level of education",s=100)
plt.savefig("abc.png")
```
